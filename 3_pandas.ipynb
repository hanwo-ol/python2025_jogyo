## 3.5. 데이터 가공 및 그룹핑

```python
df = pd.read_csv('./data/music.csv')
df.head()
```

```python
# 열 삭제
# drop() 메소드는 원본 데이터를 수정하지 않음
df.drop(columns=['sortation', 'agency'])
#df.drop(['sortation', 'agency'], axis=1)
```

```python
# 열 삭제
# drop() 메소드는 원본 데이터를 수정하지 않음
# -> 원본을 바꾸면 싶은면? 결과를 원본에 저장!
# -> 또 다른 방법은 inplace=True 옵션을 사용함!
df = df.drop(columns=['sortation', 'agency'])
# df.drop(columns=['sortation', 'agency'], inplace=True)
df.head()
```

```python
# 행 삭제
df.drop(index=range(10))
#df.drop(range(10), axis=0)
```

```python
df.head()
```

```python
# 열 이름 변경
df.columns = ['노래제목', '가수', '인원수', '발매일', '장르', '앨범유형']
df.head()
```

```python
# 발매일 변수를 datetime 형식으로 변환 : pd.to_datetime() 함수 사용
df['발매일'] = pd.to_datetime(df['발매일'])
df.dtypes
```

```python
# 새로운 열 생성
df['연도'] = df['발매일'].dt.year
df['월'] = df['발매일'].dt.month
df['일'] = df['발매일'].dt.day
df.head()
```

```python
# 데이터 그룹핑1 : 각 장르의 데이터 개수
#df.value_counts('장르')
newdata = df.groupby('장르').count()
newdata['노래제목']
```

```python
df.head(3)
```

```python
# 데이터 그룹핑2
df.groupby('장르')['인원수'].mean()
```

```python
df.groupby('장르')['인원수'].max()
```

```python
df.groupby('장르')['인원수'].min()
```

## 3.6. 결측 데이터 처리

```python
df = pd.read_csv("./data/train.csv")
df.head(2)
```

```python
# df.isna()
df.isnull()
```

```python
# Embarked 변수가 결측값인 경우만 선택
# -> Embarked 변수에 대해 결측값이 있으면 True, 없으면 False를 가지는 논리형 배열이 필요함!
mask = df['Embarked'].isnull()
df.loc[mask, :]
```

```python
# 논리형 배열에서 True의 개수는? -> 배열의 합을 구하면 됨
# 논리형은 True, False 값을 가짐 -> 연산을 적용하면? True는 1로 취급, False는 0으로 취급
arr = np.array([True, False, True, True, False])
print(arr.sum())
```

```python
df.shape
```

```python
# 각 열별 결측값 개수 확인
df.isnull().sum()
```

```python
# 결측값은 어떻게 처리해야 하는가?
# -> 특정 열에서 결측값이 많이 보이고 크게 중요하지 변수가 아니면, 해당 열을 삭제
df.drop(columns='Cabin', inplace=True)
df.head(2)
```

```python
# 결측값이 하나라도 포함된 행 삭제
# -> 결측값이 하나라도 포함된 행이 적은 경우
df.dropna(axis=0)
```

```python
# 결측값을 최빈값으로 대체
# -> Embarked 변수의 최빈값(mode)을 구해야 함
mode = df['Embarked'].mode().loc[0,]            # df['Embarked'].mode()[0]
df['Embarked'] = df['Embarked'].fillna(mode)
df.isna().sum()
```

```python
# 결측값을 평균으로 대체
mean = df['Age'].mean()
df['Age'] = df['Age'].fillna(mean)
df.isna().sum()
```
