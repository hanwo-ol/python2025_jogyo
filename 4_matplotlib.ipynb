```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 기본 그래프
plt.figure(figsize=(3, 2))
xdata = [3, 5, 7, 9]
ydata = [2, 4, 6, 8]
plt.plot(xdata, ydata)
plt.show()
```

```python
# 시각화 옵션1 : 제목
plt.figure(figsize=(3, 2))
plt.title('Line Graph')
xdata = [3, 5, 7, 9]
ydata = [2, 4, 6, 8]
plt.plot(xdata, ydata)
plt.show()
```

```python
# 시각화 옵션2 : 범례(legend)
# 두 개의 선 그래프
data1 = [1, 3, 5, 7]
data2 = [9, 7, 5, 3]

plt.figure(figsize=(10, 5))
plt.plot(data1, label='Salary', color='purple')
plt.plot(data2, label='Price', color='green')
plt.legend()
plt.show()
```

```python
# 시각화 옵션3 : 색상
plt.figure(figsize=(3, 2))
xdata = [3, 5, 7, 9]
ydata = [2, 4, 6, 8]
plt.plot(xdata, ydata, color='#92bcf7')
plt.show()
```

```python
# 시각화 옵션4 : x축 및 y축 이름
plt.figure(figsize=(3, 2))
plt.title('Axis name Properties')
xdata = [3, 5, 7, 9]
ydata = [2, 4, 6, 8]
plt.plot(xdata, ydata)
plt.xlabel('x value')
plt.ylabel('y value')
plt.show()
```

```python
# 시각화 옵션5 : 그래프 선 모양
# linestyle 속성에 선 모양을 '-', '--', '-.', ':' 설정
data1 = [1, 3, 5, 7]
data2 = [9, 7, 5, 3]

plt.figure(figsize=(3, 2))
plt.plot(data1, label='Salary', color='purple', linestyle='--')      # 'dashed'
plt.plot(data2, label='Price', color='green', linestyle=':')         # 'dotted'
plt.legend()
plt.show()
```

```python
# 시각화 옵션6 : 그림 범위 지정
xdata = [15, 25, 35, 45]
ydata = [2, 4, 6, 8]

plt.figure(figsize=(3, 2))
#plt.plot(xdata, ydata, color='m', marker='s')

# 색상, 선 모양, 마커를 한번에 설정할 수 있음
# 색상 : r g b c m y k w
# 선 모양 : - -- -. :
# 마커 : + , . o v ^ s 
plt.plot(xdata, ydata, ':oy')
    
plt.xlim(10, 50)
plt.ylim(0, 10)
plt.show()
```

```python
data = pd.read_csv('./data/iris.csv')
print(data.shape)
print(data.dtypes)
```

```python
data.columns = ['SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth', 'Species']
data.columns
```

```python
plt.figure(figsize=(4, 2))
plt.scatter(data['PetalLength'], data['PetalWidth'], color='orange', s=5)
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
```

```python
data['Species'].unique()
```

```python
colors = data['Species'].map({'Iris-setosa': 'red',
                              'Iris-versicolor': 'green',
                              'Iris-virginica': 'blue'})
```

```python
plt.figure(figsize=(4, 2))
plt.scatter(data['PetalLength'], data['PetalWidth'],
            c=colors, s=5)
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
```

```python
plt.figure(figsize=(4, 2))
plt.scatter(data['PetalLength'], data['SepalLength'],
            c=colors, s=5)
plt.xlabel('Petal Length')
plt.ylabel('Sepal Length')
```

```python
mask = data['Species'] == 'Iris-setosa'
setosa = data.loc[mask, :]
setosa
```

```python
plt.figure(figsize=(5, 3))
plt.hist(setosa['SepalWidth'], bins=8, color='pink', edgecolor='black')
plt.xlabel('Sepal Width')
plt.ylabel('frequency')
plt.show()
```
